1.9 Data-Ingestion:

Ingestion and RAG can occur at the same time!

Paralleling is efficient, especially when the data you need to ingest is HUGE

```python
#This is for querying the SQLite database, while the ingestion is taking place.
#To illustrate how we can ingest data into the database, and ALSO query it for the RAG purpose.
#sqlite_index is similar to "index" in rag_helper.py.
 
from sqlitesearch import TextSearchIndex

sqlite_index = TextSearchIndex(
    text_fields=["question", "section", "answer"],
    keyword_fields=["course"],
    db_path="faq.db"
)
```

```python
sqlite_index.count()
sqlite_index.search('how do I join the course')
```

Now we can RAG as before, BUT we use sqlite_index, instead of index from minsearch

```python
from rag_helper import RAGBase
from openai import OpenAI
from dotenv import load_dotenv

openai_client = OpenAI()
load_dotenv()
assistant=RAGBase(sqlite_index,openai_client)
```

```python
assistant.rag("How do I get a certificate?")
```

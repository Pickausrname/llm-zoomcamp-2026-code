1.9 Data Ingestion (into a SQLITE DB)

CONCEPT:

In module 1.4 we retrieved FAQ data from the course URL's JSON page, then stored it into a variable "document".

"Document" disappears once we close the notebook. Not very efficient if we retrieve it everytime. We want to store it in a DB, so we dont have to fetch it from the web everytime.

STEPS:

-We still load the web contents into "documents" as before.

-We then filter it down to the "llm-zoomcamp" contents.

-Then we index it using the SQLITEsearch API "TestSearchIndex" - just like what we did with minsearch in module 1.5

-Finally we save it to a sqlite DB.

```python
from ingest import load_faq_data
documents= load_faq_data()
print(f"LLM Zoomcamp: {len(documents)} documents")
```

```python
docs_llm = [doc for doc in documents if doc["course"] == "llm-zoomcamp"]
print(f"LLM Zoomcamp: {len(docs_llm)} documents")
print(f"{docs_llm[1]}")
```
